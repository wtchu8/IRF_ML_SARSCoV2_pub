```python
import os
import pandas as pd
import yaml
import pickle
import sys
```

```python
# Local packages
sys.path.append('../src')
from ninetynine import ninetynine
```

```python
#Retreive local data path
with open('../config/paths.yaml','r') as file:
    paths_list = yaml.safe_load(file)
    PATH = os.path.abspath(paths_list['PATH'])

# Retreive s_list
with open('../config/s_list.yaml','r') as file:
    s_list = yaml.safe_load(file)['s_list']
```

```python
# Getting back the objects:
with open('../config/lists_radio.pkl', 'rb') as f:
    id_cols_radio,var_cols_radio,time_points,delta_time = pickle.load(f)

# Open cleaned data
data_radio = pd.read_csv(os.path.join(PATH,'tables','data_radio_delta.csv'))
data_radio.head()
```

```python
# Getting back the objects:
with open('../config/lists_path.pkl', 'rb') as f:
    id_cols_path,var_cols_path = pickle.load(f)

# Open cleaned data
data_path = pd.read_csv(os.path.join(PATH,'tables','data_path_delta.csv'))
data_path.head()
```

```python
# Getting back the objects:
with open('../config/lists_imm.pkl', 'rb') as f:
    id_cols_imm,var_cols_imm = pickle.load(f)

# Open cleaned data
data_imm = pd.read_csv(os.path.join(PATH,'tables','data_imm_delta.csv'))
data_imm.head()
```

```python
id_cols = id_cols_radio + id_cols_path + id_cols_imm
var_cols = var_cols_radio + var_cols_path + var_cols_imm
```

```python
data_merge = data_path.set_index(['Subject','StudyDay']).join(data_radio.set_index(['Subject','StudyDay']),lsuffix='_path', rsuffix='_radio').reset_index()
data_merge
```

```python
data_merge = data_merge.set_index(['Subject','StudyDay']).join(data_imm.set_index(['Subject','StudyDay']),rsuffix='_imm').reset_index()
data_merge
```

```python
# Check if the two datasets have matching values
ninetynine(not data_merge['Class_radio'].equals(data_merge['Class_path']),'mis-matched class labels')
ninetynine(not data_merge['Class_radio'].equals(data_merge['Class']),'mis-matched class labels')
#data_merge.loc[data_merge.Class_radio != data_merge.Class_path,['Class_path','Class_radio']]
```

```python
# Create a single Class field
data_merge['Class_name'] = data_merge['Class_path']

# Enumerate Class
data_merge.loc[data_merge["Class_name"] == "Infected","Class"] = 1
data_merge.loc[data_merge["Class_name"] == "Mock","Class"] = 0
```

#### Check and save results

```python
data_merge = data_merge.loc[data_merge.Subject.isin(s_list)]
print('Subject count: ',len(data_merge.Subject.unique()))

#Write status message, change this anytime this part of the analysis changes
status='SexOriginBal'
with open(os.path.join('..','config','analysis_status','t_merge.txt'),'w') as out_file:
    out_file.write(status)
```
