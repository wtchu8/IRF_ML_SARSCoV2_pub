```python
import os
import sys
import pandas as pd
import yaml
```

```python
import scipy.stats
```

```python
#Retreive local data path
with open('../config/paths.yaml','r') as file:
    paths_list = yaml.safe_load(file)
    PATH = os.path.abspath(paths_list['PATH'])
```

```python
data = pd.read_csv(os.path.join(PATH,'tables','data_merge_noDelta.csv'))
data.groupby(['StudyDay','Class_name']).describe()
```

```python
data_clinScore = pd.read_csv(os.path.join(PATH,'raw','Cage-side_scores_v2.csv'))

data_clinScore = data_clinScore.pivot_table(index = 'Subject', values = 'Clinical_score', columns = 'StudyDay')
data_clinScore.columns = data_clinScore.columns.astype(str)
```

```python
data_clinScore
```

```python
# Calculate the change from pre-exposure
data_clinScore['2_delta'] = data_clinScore['2'] - data_clinScore['-7']
data_clinScore['4_delta'] = data_clinScore['4'] - data_clinScore['-7']
data_clinScore['6_delta'] = data_clinScore['6'] - data_clinScore['-7']
data_clinScore['max'] = data_clinScore[['2','4','6']].max(axis=1)
```

```python
# get the baseline age in years
data['Age_bl_y'] = (data.loc[data.StudyDay==2,'Age_d']-2)/365
```

```python
data_clin_score = data[['Subject','Class_name','SEX','COUNTRY_ORIGIN','Age_bl_y']].dropna().drop_duplicates(subset=None).set_index('Subject').join(data_clinScore[['max']])
data_clin_score
```

```python
data_clin_score.groupby('Class_name').describe()
```

```python
def welch_dof(x,y):
        dof = (x.var()/x.size + y.var()/y.size)**2 / ((x.var()/x.size)**2 / (x.size-1) + (y.var()/y.size)**2 / (y.size-1))
        print(f"Welch-Satterthwaite Degrees of Freedom= {dof:.4f}")
```

```python
scipy.stats.ttest_ind(data_clin_score.loc[data_clin_score.Class_name == 'Infected','max'],data_clin_score.loc[data_clin_score.Class_name == 'Mock','max'], equal_var=False)
```

```python
welch_dof(data_clin_score.loc[data_clin_score.Class_name == 'Infected','max'], data_clin_score.loc[data_clin_score.Class_name == 'Mock','max'])
```

```python
scipy.stats.ttest_ind(data_clin_score.loc[data_clin_score.Class_name == 'Infected','Age_bl_y'],data_clin_score.loc[data_clin_score.Class_name == 'Mock','Age_bl_y'], equal_var=False)
```

```python
welch_dof(data_clin_score.loc[data_clin_score.Class_name == 'Infected','Age_bl_y'], data_clin_score.loc[data_clin_score.Class_name == 'Mock','Age_bl_y'])
```

```python
data_weight = data_weight.set_index('Animal ID')
```

```python
data_weight['BL'] = data_weight[['D-11','D-7']].mean(axis=1)
data_weight['delta_2'] = data_weight.D2 - data_weight.BL
data_weight['delta_4'] = data_weight.D4 - data_weight.BL
data_weight['delta_6'] = data_weight.D6 - data_weight.BL
```

```python
data_weight['delta_mean'] = data_weight[['delta_2','delta_4','delta_6']].mean(axis=1)
```
