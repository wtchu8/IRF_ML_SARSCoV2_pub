```python
import os
import pandas as pd
import yaml
import pickle
```

```python
# Getting back the objects:
with open('../config/lists_radio.pkl', 'rb') as f:
    id_cols_radio,var_cols_radio,time_points,delta_time = pickle.load(f)

# Open cleaned data
data_radio = pd.read_csv(os.path.join(PATH,'tables','data_radio.csv'))
data_radio = data_radio.loc[data_radio.StudyDay.isin(['pre','2','4','6','8'])]
data_radio.head()
```

```python
# Getting back the objects:
with open('../config/lists_path.pkl', 'rb') as f:
    id_cols_path,var_cols_path = pickle.load(f)

# Open cleaned data
data_path = pd.read_csv(os.path.join(PATH,'tables','data_path.csv'))
data_path = data_path.loc[data_path.StudyDay.isin(['pre','2','4','6','8'])]
data_path.head()
```

```python
# Getting back the objects:
with open('../config/lists_imm.pkl', 'rb') as f:
    id_cols_imm,var_cols_imm = pickle.load(f)

# Open cleaned data
data_imm = pd.read_csv(os.path.join(PATH,'tables','data_imm.csv'))
data_imm = data_imm.loc[data_imm.StudyDay.isin(['pre','2','4','6','8'])]
data_imm.head()
```

```python
def ninetynine(condition, problem):
    # condition: True if there is a problem
    if condition:
        print('I\'ve got 99 problems, and this is the 100th: ' + problem)
    else:
        print('I\'ve got 99 problems, but ' + problem + ' is not one')
```

```python
# Check if the two datasets have matching values
ninetynine(not data_merge['Class_radio'].equals(data_merge['Class_path']),'mis-matched class labels')
ninetynine(not data_merge['Class_radio'].equals(data_merge['Class']),'mis-matched class labels')
#data_merge.loc[data_merge.Class_radio != data_merge.Class_path,['Class_path','Class_radio']]
```
